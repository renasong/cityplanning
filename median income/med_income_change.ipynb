```python
import pandas as pd
import re
```

```python
df2010=pd.read_csv('2010_med_income.csv')
df2018=pd.read_csv('2018_med_income.csv')
```

```python
df2010.drop(df2010.columns[[0]], axis=1, inplace=True)
df2018.drop(df2018.columns[[0]], axis=1, inplace=True)
```

```python
m2010=[]

for d in df2010['Median income (dollars)  Estimate  Households']:
    if d==r'-':
        m2010.append(0)
    else:
        m2010.append(int(d))
```

```python
m2018=[]

for d in df2018['Median income (dollars)  Estimate  Households']:
    if d==r'-':
        m2018.append(0)
    else:
        m2018.append(int(d))
```

```python
new=df2010
```

```python
new['2010']=m2010
new['2018']=m2018
```

```python
new=new[['Geographic Area Name','2010','2018']]
```

```python
new['change']=new['2018']-new['2010']
```

```python
tracts=[]

for d in new['Geographic Area Name']:
    if re.findall(r'(\d\d\d\d.\d\d|\d\d\d\d.\d|\d\d\d\d|\d\d\d.\d\d|\d\d\d.\d|\d\d\d)',d):
        tracts.append(float(re.findall(r'(\d\d\d\d.\d\d|\d\d\d\d.\d|\d\d\d\d|\d\d\d.\d\d|\d\d\d.\d|\d\d\d)',d)[0]))
    else:
        tracts.append('')
    
new['tract']=tracts
new['tract']=new['tract'].astype(str)
```

```python
sf=pd.read_csv('sf polygons-Copy1.csv')
```

```python
sf['NAME']=sf['NAME'].astype(str)
```

```python
df2=new.merge(sf[['NAME','geometry']],left_on='tract',right_on='NAME')
df2=df2.drop(columns=['NAME'])
```
