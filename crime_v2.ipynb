```python
import geopandas as geo
import pandas as pd
import shapely.wkt
from shapely.geometry import Point, Polygon
import re
import time
import datetime
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/Users/renasong/Downloads/Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv')
```

```python
sf=pd.read_csv('sf polygons.csv')
```

```python
polys = [shapely.wkt.loads(st) for st in sf['geometry']]
```

```python
points=[]
for index, row in df.iterrows():
    points.append(Point(row['X'],row['Y']))
```

```python
years=[re.findall(r'\d{4}',date)[0] for date in df['Date']]
df['Year']=years
```

```python
tracts=[]
for point in points:
        count=0
        for poly in polys:
            if poly.contains(point):
                tracts.append(poly)
                count+=1
        if count<1:
            tracts.append(0)
```

```python
df['Tract']=tracts
```

```python
df['Date']=[datetime.datetime.fromtimestamp(time.mktime(datetime.datetime.strptime(s, "%m/%d/%Y").timetuple())) for s in df['Date']]
```

```python
df['Tract']=[str(p) for p in df['Tract']]
```

```python
df=pd.merge(df,sf[['geometry','NAME']],left_on='Tract',right_on='geometry')
```

```python
df=df.drop(columns=['Tract'])
```
